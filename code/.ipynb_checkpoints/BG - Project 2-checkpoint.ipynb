## Problem statement

Using the Ames Housing dataset, can we build a regression model that is predictive of home value?

## Executive Summary

Key Findings:
- Many features had a correlation on sale price, even after Lasso regularization.
- Intuition only gets us so far with this data set - many strong correlations come from surprising features.

After performing exploratory data analysis and munging, we obtained a set of features that appear to have strong predictive capability for our target variable. Using scaling and regularization, we controlled for colinearity. By using a hold out set and cross validation, we controlled for overfitting on our training data, and ultimately achieved an $R^2$ score of `0.93`.

## 1. Importing and Examining Data

```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV
from sklearn.preprocessing import StandardScaler, PowerTransformer
from sklearn.metrics import r2_score

%matplotlib inline

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)

import warnings
warnings.filterwarnings("ignore", category=FutureWarning) 
```

```python
# Reading the 'train.csv' datafile as a pandas DataFrame
train_df = pd.read_csv('../datasets/train.csv')
test_df = pd.read_csv('../datasets/test.csv')

# Creating our submissions dataframe, preserving the 'Id' column
test_predictions = test_df[['Id']]
```

```python
# Examining the columns in our training data
train_df.info()
```

> There are some missing values. According to our data dictionary, many ordinal columns are coded as strings.

```python
# Function to clean up column names
def fix_columns(data):
    data.columns = [col.lower().replace(' ', '_') for col in data.columns]
    return data
```

```python
# Calling the column cleaning function
train_df = fix_columns(train_df)
test_df  = fix_columns(test_df)
```

```python
# Dropping columns with little to no data
columns = ['alley', 'pool_qc', 'fence', 'misc_feature', 'garage_yr_blt']
train_df.drop(columns, axis = 1, inplace = True)
test_df.drop(columns, axis = 1, inplace = True)
```

## 4. Cleaning missing data

Looking at the data dictionary, it appears most of our missing values correspond with features of the home that were nonexistent (basements, garages, fireplaces, etc). It makes sense to impute these values with 'NA' or 0 depending on the column.

```python
# Function to fill missing values
def fill_missing(data):
    
    # Missing values in 'fireplace_qu' seem to correspond with homes with no fireplace, so adding a new category:
    data['fireplace_qu'].fillna('NA', inplace = True)
    
    # Filling missing 'lot_frontage' with the mean of the column:
    lot_frontage_mean = data['lot_frontage'].mean()
    data['lot_frontage'].fillna(lot_frontage_mean, inplace = True)
    
    # Assuming missing 'mas_vnr_type' and 'mas_vnr_area' values are 'None' and 0 respectively:
    data['mas_vnr_type'].fillna('NA', inplace = True)
    data['mas_vnr_area'].fillna(0, inplace = True)
    
    # Assuming missing 'bsmt' columns refer to observations with no basement:
    data['bsmt_qual'].fillna('NA', inplace = True)
    data['bsmt_cond'].fillna('NA', inplace = True)
    data['bsmt_exposure'].fillna('NA', inplace = True)
    data['bsmtfin_type_1'].fillna('NA', inplace = True)
    data['bsmtfin_type_2'].fillna('NA', inplace = True)
    
    # Assuming missing 'garage' columns refer to observations with no garage:
    data['garage_type'].fillna('NA', inplace = True)
    data['garage_finish'].fillna('NA', inplace = True)
    data['garage_qual'].fillna('NA', inplace = True)
    data['garage_cond'].fillna('NA', inplace = True)

    return data
```

```python
# Applying our function to both the train and test data sets
train_df = fill_missing(train_df)
test_df = fill_missing(test_df)
```

```python
# Drop the few remaining missing rows
train_df = train_df.dropna()
```

## 5. Data transformations

From the data dictionary, quite a few of the columns are ordinal strings. We'll convert these to numeric values on a range that keeps the intent of the original ordinal column.

```python
# Transforming ordinal values to numeric:
def ordinal_data(data):
    data['exter_qual'] = data['exter_qual'].map({'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})
    data['exter_cond'] = data['exter_cond'].map({'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})
    data['bsmt_qual'] = data['bsmt_qual'].map({'NA':0, 'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})
    data['bsmt_exposure'] = data['bsmt_exposure'].map({'NA':0, 'No': 10, 'Mn': 12, 'Av': 13, 'Gd': 20})
    data['bsmtfin_type_1'] = data['bsmtfin_type_1'].map({'NA':0, 'Unf': 10, 'LwQ': 12, 'Rec': 13, 'BLQ': 20, 'ALQ': 25, 'GLQ': 30})
    data['heating_qc'] = data['heating_qc'].map({'NA':0, 'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})
    data['kitchen_qual'] = data['kitchen_qual'].map({'NA':0, 'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})
    data['garage_finish'] = data['garage_finish'].map({'NA':0, 'Unf': 10, 'RFn': 12, 'Fin': 20})
    data['fireplace_qu'] = data['fireplace_qu'].map({'NA':0, 'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})
    data['garage_cond'] = data['garage_cond'].map({'NA':0, 'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})
    data['garage_qual'] = data['garage_qual'].map({'NA':0, 'Po': 10, 'Fa': 12, 'TA': 13, 'Gd': 20, 'Ex': 30})

    return data
```

```python
# Applying our ordinal transformation function to both the training and test datasets
train_df = ordinal_data(train_df)
test_df = ordinal_data(test_df)
```

## 6. Exploratory Data Analysis

Let's look at our correlation matrix.

```python
# Creating the correlation matrix and saving it to 'corr'
corr = train_df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(25, 25))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap='coolwarm', center=0, vmax = 1,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}).set_title('Correlation Heatmap');
```

> There are quite a few variables strongly correlated with saleprice. There are also quite a few variables that exhibit colinearity.

## 7. Creating Dummy Variables

We have quite a few categorical variables that may be correlated with sale price. We'll create dummy variables using all of them, ensure the train and test datasets are consistent, and use the variables most strongly correlated with saelprice for our final features.

```python
# Create dummy variables:
dummies_train = pd.get_dummies(train_df, drop_first = True)
dummies_test = pd.get_dummies(test_df, drop_first = True)
```

```python
# Taking only features with an absolute correlation with saleprice of 0.05 or greater:
corr_dummies = dummies_train.corr().abs()[['saleprice']].sort_values(by = 'saleprice', ascending = False)
features = corr_dummies[(corr_dummies['saleprice'] > 0.05) & 
                        (corr_dummies['saleprice'] < 1.0)][['saleprice']].sort_values(by = 'saleprice', 
                                                                                      ascending = False).index
```

```python
# Plotting the resulting features as correlated with saleprice
fig, ax = plt.subplots(figsize = (2,30))
sns.heatmap(corr_dummies[['saleprice']].sort_values('saleprice', ascending = False).head(145),  
            cmap = 'coolwarm',
            center = 0, 
            vmax = 1, 
            annot = True).set_title('Absolute Correlation with Sale Price');
```

```python
# How many features did we end up with?
features.shape
```

> On first glance this appears to be too many. We'll use Lasso later to minimize the coefficients to narrow down to which features actually impact our predictions.

```python
# Creating missing dummy columns in the test dataset
columns = ['garage_qual_Fa',
           'ms_zoning_C (all)',
           'bsmt_cond_Fa', 
           'functional_Sal', 
           'heating_Wall', 
           'roof_matl_CompShg', 
           'bsmt_cond_Po',
           'condition_2_Feedr',
           'electrical_Mix']

for col in columns:
    dummies_test[col] = 0
```

## 8. Building X and y, transforming and scaling features

```python
# Setting up our X (train and test feature matrices) and y (train target vector)
X_train = dummies_train[features]
X_test  = dummies_test[features]
y_train = dummies_train['saleprice']
```

> Similar to how we took the log of saleprice at the beginning of this notebook, PowerTransformer will transform our features to better reflect linear relationships with our target.

```python
# Scaling our features
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc  = ss.transform(X_test)
```

```python
# Using PowerTransformer to attempt to make our data more Gaussian
pt = PowerTransformer()
pt.fit(X_train_sc)
X_train_pt = pt.transform(X_train_sc)
X_test_pt  = pt.transform(X_test_sc)
```

> We apply StandardScaler to standardize each feature to +/- standard deviations from its mean, to weight each feature equally.

## 9. Constructing and scoring the model

We use LassoCV to minimize effects of colinearity between features. This will heavily penalize our coefficients that do not contribute to the predictive nature of the model.

```python
# Finding the best alpha for our Lasso model
lasso = LassoCV()
lasso.fit(X_train_pt, y_train)
alpha = lasso.alpha_

# Zeroing in on the best alpha
lasso = LassoCV(alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, 
                          alpha * .85, alpha * .9, alpha * .95, alpha, alpha * 1.05, 
                          alpha * 1.1, alpha * 1.15, alpha * 1.25, alpha * 1.3, alpha * 1.35, 
                          alpha * 1.4], 
                max_iter = 50000, cv = 10)

# Cross validating with 5 folds
cross_val_score(LassoCV(), X_train_pt, y_train, cv = 5).mean()
```

> The mean of our R2 scores over 5 folds is 0.92

```python
# Plotting the top 10 features
features_df = pd.DataFrame(X_train_pt, columns = features)
features_df.loc[:,'overall_qual':'year_built'].hist(figsize = (20,20));
```

> Our top features appear roughly normally distributed with the exception of year_built

```python
# Fitting and scoring our model
lasso.fit(X_train_pt, y_train)
lasso.score(X_train_pt, y_train)
```

> Our final $R^2$ score is 0.933, which is not out of line with our cross-validated scores, which seems to imply we're not overfitting.

```python
# How many features are we using after Lasso?
new_df = pd.DataFrame(lasso.coef_, index = features).abs().sort_values(by = 0, ascending = False)
new_df[new_df[0] > 0].shape
```

> 105 features had coefficients above 0 after Lasso. This implies that they have predictive qualities for our target.

## 10. Generating the Output File

```python
# Generating predictions from our model
pred = lasso.predict(X_test_pt)
```

```python
# Writing our predictions back to the test_predictions dataframe and converting the target back to normal units
test_predictions['SalePrice'] = np.expm1(pred)
```

```python
# Validating the format of our predictions dataframe
test_predictions.head()
```

```python
# Validating the shape of our predictions dataframe
test_predictions.shape
```

```python
# Writing the resulting dataframe to 'predictions.csv'
test_predictions.to_csv('../datasets/predictions.csv', index = False)
```

## 11. Validating through a holdout set

```python
# Generating a holdout set to validate our assumptions
X_h_train, X_h_test, y_h_train, y_h_test = train_test_split(X_train_pt, y_train, random_state = 21)
```

> By regenerating the above holdout set with various `random_state` values, we can cross validate our score and check for overfitting.

```python
# Generating the mean of cross validated scores on our holdout set, with 5 folds
cross_val_score(LassoCV(), X_h_train, y_h_train, cv = 5).mean()
```

```python
# Fitting and scoring a new model for the holdout set
lasso2 = LassoCV()
lasso2.fit(X_h_train, y_h_train)
lasso2.score(X_h_train, y_h_train)
```

> This is in line with our $R^2$ score above

```python
# Calculating our predicted prices and the intercept
y_hat = lasso2.predict(X_h_test)
intercept = lasso2.intercept_
```

```python
# Plotting predicted vs actual price
fig, ax = plt.subplots(figsize = (10,10))
sns.regplot(np.expm1(y_h_test), np.expm1(y_hat))
ax.set_xlabel('Predicted Price')
ax.set_ylabel('Actual Price')
ax.set_title('Predicted vs Actual Sale Price');
```

> The plot of predicted vs actual price follows nicely with our regression model. There is a good distribution of over - vs - under predictions and the residuals do not appear to share a linear relationship with our predictions.

```python
# Calculating and plotting the distribution of residuals
residuals = y_h_test - y_hat
fig, ax = plt.subplots(figsize = (8,5))
sns.distplot(residuals)
ax.set_title('Distribution of residuals')
ax.set_xlabel('Difference in predicted vs actual price');
```

> A few outliers, but mostly normal distribution.

## 12. Conclusion

After performing exploratory data analysis and munging, we obtained a set of features that appear to have strong predictive capability for our target variable. Using scaling and regularization, we controlled for colinearity. By using a hold out set and cross validation, we controlled for overfitting on our training data, and ultimately achieved an $R^2$ score of `0.93`.

## Bonus - Gradient Boosting Regressor

Out of curiosity, we'll try a Gradient Boosting Regression Tree to compare the results with Lasso.

```python
# Importing
from sklearn.ensemble import GradientBoostingRegressor 
from sklearn.model_selection import GridSearchCV

# Setting the parameter grid to search over
param_grid={'n_estimators':[100], 
            'learning_rate': [0.1,0.05,0.02], 
            'max_depth':[4,6], 
            'min_samples_leaf':[3,5,9,17], 
            'max_features':[1.0,0.3,0.1]
           }
n_jobs = 3

# Fitting a Gradient Boosting Regressor model, choosing the best hyperparameters from param_grid
grid = GridSearchCV(GradientBoostingRegressor(), param_grid = param_grid, n_jobs = n_jobs)
grid.fit(X_train_pt, y_train)
```

```python
# What was the best result?
grid.best_estimator_
```

```python
# Generating predictions on the scaled test data
gbr = grid.best_estimator_
pred2 = gbr.predict(X_test_sc)
```
